```python
import numpy as np
import pandas as pd
import pylab as pl
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import subprocess
import pydot 
from IPython.display import Image
from IPython.core.display import HTML
```

```python
df = pd.read_csv("ORGANICS.csv", encoding="cp1251")
print(df.shape) 
print(df.info())
```

```python
#1) Удаление Nan
df.dropna()
```

```python
#2) Замена Nan усредненными значениями
df = df.fillna(df.mean())
```

```python
#3) Замена Nan наиболее часто встречаемыми значениями
df = df.fillna(df.mod)
```

```python
x = df[['DemAffl',  'DemAge', 'DemCluster', 'PromSpend', 'PromTime']]
y = df.values[::,11]
y = y.astype('int')
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
tree=DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
```

```python
export_graphviz(tree, out_file='result/Organics_tree.dot', 
               rounded = True, proportion = False, 
                precision = 2, filled = True)
```

```python
(graph, ) = pydot.graph_from_dot_file('result/Organics_tree.dot')

graph.write_png('result/Organics_tree.png')
```

```python
tree.score(X_train, y_train) 
```

```python
tree.score(X_test,y_test) 
```
