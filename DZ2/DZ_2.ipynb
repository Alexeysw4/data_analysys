```python
#1 Получите таблицу со студентами с четными номерами.
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('math_students.csv', delimiter=',')

data[::2]
```

```python
#2 Получите таблицу со студентами, имеющими четное число пропусков.
filter = data['absences'] % 2 == 0
data.loc[filter]
```

```python
# 3 Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?
less = data[data['absences'] <= 10]
less = less.G3.mean()
print('средний балл у тех у кого менее 10:\n', less)
more = data[data['absences'] > 10]
more = more.G3.mean()
print('средний балл у тех у кого более 10:\n', more)

if less < more:
    print('Нет, больше пропусков - выше балл')
else:
    print('Да, больше пропусков - ниже балл')
```

```python
#4 Как вы думаете, какие еще признаки могут повлиять на итоговую оценку? Найдите закономерности, аналогичные рассмотренным.
yes = data[data['activities'] == 'yes']
yes = yes.G3.mean()
print('средний балл у людей с дополнительной активность0:\n', yes)
no = data[data['activities'] == 'no']
no = no.G3.mean()
print('средний балл без дополнительной активности:\n', no)

if no < yes:
    print('Есть доп активность - выше балл')
else:
    print('Нет доп активности - ниже балл')
```

```python
#5 Как вы думаете, если при школе откроется общежитие, насколько актуально это будет для студентов математического класса?
#Я посмотрю на количество иногородних студентов (filter_address) 
#и на количество студентов поступающих из-за хорошего курса(filter_reason2)
#или престижа учебного заведения (filter_reason1). если у заведения 
#высокий престиж, то велик шанс что в него начнут приезжать люди
#из других стран/городов и им нужено будет общежитие.
 
# Если коэфы filter_reason1 и filter_reason2 будут достаточно высоки,
#то это может оказаться хорошим вложением



filter_address = data['address'] == 'R'
filter_reason1 = data['reason'] == 'reputation' 
filter_reason2 = data['reason'] == 'course'
a = data.loc[filter_address].shape 
b = data.shape
f1 =a[0] / b[0]
print('Процент сельских студентов:\n', f1)

c = data.loc[filter_reason1 | filter_reason2].shape 
d = data.shape
f2 = c[0] / d[0]
print('Процент положительных отзывов:\n', f2)

f3 = f1 + f2
f2 = c[0] / d[0]
print('Сумма коэфов:\n', f3)

if f3 > 0.7:
    print('\nКоэф высок, нужно общежитие')
else:
    print('\nКоэф низок, постройка общежития может оказаться экономически не выгодна')
```

```python
#6 Найдите в данных закономерности с помощью группировки.
d1 = data.groupby('Dalc')
print(d1.G3.mean(), '\n\nСамые высокие оценки у тех кто в будни либо пьет очень мало, либо у тех кто пьет очень много :3')
```

```python
data = pd.read_csv('math_students.csv', delimiter=',')
print(data.info, '\n')
print(data.shape, '\n')
print(data.columns, '\n')

if data.isnull().any().any() == True:
    data=data.fillna(data.median())
    
print(data.describe(), '\n')
print(data, '\n')
data.to_csv('new_math_students.csv')
```
